```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from file
frequencies = []
real_parts = []
imag_parts = []

with open('Impedance vs Freq 4.txt', 'r') as file:
    for line in file:
        if not line.startswith('%') and line.strip():  # Skip comments and empty lines
            parts = line.split()
            if len(parts) >= 2:
                try:
                    freq = float(parts[0])
                    impedance = complex(parts[1].replace('i', 'j'))  # Convert to Python complex format
                    frequencies.append(freq)
                    real_parts.append(impedance.real)
                    imag_parts.append(impedance.imag)
                except (ValueError, IndexError):
                    continue

# Convert to numpy arrays for easier manipulation
frequencies = np.array(frequencies)
real_parts = np.array(real_parts)
imag_parts = np.array(imag_parts)

# Create the plot
plt.figure(figsize=(12, 6))

# Plot real part
plt.subplot(1, 2, 1)
plt.plot(frequencies, real_parts, 'b-', label='Real Part')
plt.xlabel('Frequency (GHz)')
plt.ylabel('Impedance (S)')
plt.title('Real Part of Admittance')
plt.grid(True)
plt.legend()

# Plot imaginary part
plt.subplot(1, 2, 2)
plt.plot(frequencies, imag_parts, 'r-', label='Imaginary Part')
plt.xlabel('Frequency (GHz)')
plt.ylabel('Impedance (S)')
plt.title('Imaginary Part of Admittance')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.savefig('impedance.png')
plt.show()
```
